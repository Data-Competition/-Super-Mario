```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data/mushrooms.csv')

data.drop('stalk-shape', axis=1, inplace=True)
```

```python
import collections


def stack_bar_processing(d_):
    dists = collections.OrderedDict()
    
    for c in d_.columns[1:]:
        print("Column : %s" % c)
        keys = d_[c].unique()

        d_k = collections.OrderedDict()
        for k in keys:
            d = pd.concat([d_['class'], d_[c]], axis=1)

            e = d.loc[(d['class'] == 'e') & (d_[c] == k)]
            p = d.loc[(d['class'] == 'p') & (d_[c] == k)]

            e_len, p_len = len(e), len(p)
            total = e_len + p_len

            print("[*] key %s : %d/%d (%.3f/%.3f) (total %d)" % 
                  (k, e_len, p_len, (e_len/total), (p_len/total), total))
            d_k.update({k: [e_len, p_len]})
        dists[c] = d_k
    return dists

dists = stack_bar_processing(data)
```

```python
import os


if not os.path.exists('./stack_bar/'):
    os.mkdir("./stack_bar/")

def stack_bar(d, name="", title=None):
    e_data = [100. * x / (x + y) for x, y in d.values()]
    p_data = [100. * y / (x + y) for x, y in d.values()]
    
    plt.figure(figsize=(7, 5))
    
    width = .85
    e_bar = plt.bar(d.keys(), e_data, label='edible', color='#b5ffb9', edgecolor='white', width=width)
    p_bar = plt.bar(d.keys(), p_data, label='poison', bottom=e_data, color='#f9bc86', edgecolor='white', width=width)
    bars = [e_bar, p_bar]
    
    for axis in bars:
        for bar in axis:
            w, h = bar.get_width(), bar.get_height()
            plt.text(bar.get_x() + w / 2, bar.get_y() + h / 2, 
                     "{:.1f}".format(h),
                     ha="center", va="center")

    xtick_name = [ " %s (%d) " % (k, x + y) for k, [x, y] in d.items()]
    plt.xticks(np.arange(len(xtick_name)), xtick_name)
    plt.xlabel("Feature : %s" % name)
    plt.ylabel("Percentage (%)")
        
    plt.grid()
    plt.legend()
    plt.tight_layout()
    
    if title:
        plt.title(title)
        
        if not os.path.exists("./stack_bar/%s/" % title):
            os.mkdir("./stack_bar/%s" % title)
        
        plt.savefig('./stack_bar/%s/%s.png' % (title, name))
    else:
        plt.savefig('./stack_bar/%s.png' % name)
        
    plt.show()


for k, v in dists.items():
    stack_bar(v, k)
```

```python
## only for 'odor->n'
odor_data = data.loc[data['odor'] == 'n']
odor_data = stack_bar_processing(odor_data)

## stack bar
for k, v in odor_data.items():
    stack_bar(v, k, "odor_n")
```
