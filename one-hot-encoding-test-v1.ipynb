```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

os.chdir('/Users/joungwoongrhee/workspace/naver18/Super-Mario')
os.getcwd()

df = pd.read_csv('./data/mushrooms.csv')
df.head()
```

```python
cols = df.columns

category_list = []
for col in cols:
    category_list.append(df[col].unique())
    
category_list
```

```python
len(category_list)
```

```python
# unique는 sorting이 안됨....
category_list_sorted = []
for i in range(len(category_list)):
    tmp_list = []
    for j in range(len(category_list[i])):
        tmp_list.append(category_list[i][j])
    
    tmp_list.sort()
    category_list_sorted.append(tmp_list)
    
category_list_sorted
```

```python
[i for i in range(2)]
```

```python
category_list_sorted_num = []

for one_col in category_list_sorted:
    tmp = [i for i in range(len(one_col))]
    unit = []
    for j in tmp:
        unit.append([j])
    category_list_sorted_num.append(unit)
category_list_sorted_num
```

```python
print(category_list_sorted_num[0])
print(len(category_list_sorted_num[0]))
```

```python
one_hot_list = []
for one_col in category_list_sorted_num:
    if len(one_col) == 1:
        one_hot_list.append(np.array([[1]], dtype=float))
    elif len(one_col) == 2:        
        one_hot_list.append(np.array([[1,0],[0,1]], dtype=float))
    else :
        ohe.fit(one_col)
        ohe.n_values_, ohe.feature_indices_, ohe.active_features_
        tmp = ohe.transform(one_col).toarray()
        one_hot_list.append(tmp)

one_hot_list
```

```python
def changer_by_dict(value, col_name):
    key = col_name + '_' + value
    return one_hot_dict[key] 
```

```python
import copy
df_test = copy.deepcopy(df)

df_test['class'] = df_test['class'].apply(changer_by_dict, col_name='class')
df_test.head()
```

```python
cols = df.columns
for col in cols:
    df[col] = df[col].apply(changer_by_dict, col_name=col)

df.head()
```
